```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import Adam

# Define AlexNet model
def create_alexnet_model(input_shape, num_classes):
    model = Sequential()

    # Layer 1
    model.add(Conv2D(96, (11, 11), strides=(4, 4), activation='relu', input_shape=input_shape))
    model.add(MaxPooling2D(pool_size=(3, 3), strides=(2, 2)))

    # Layer 2
    model.add(Conv2D(256, (5, 5), activation='relu'))
    model.add(MaxPooling2D(pool_size=(3, 3), strides=(2, 2)))

    # Layer 3
    model.add(Conv2D(384, (3, 3), activation='relu'))

    # Layer 4
    model.add(Conv2D(384, (3, 3), activation='relu'))

    # Layer 5
    model.add(Conv2D(256, (3, 3), activation='relu'))
    model.add(MaxPooling2D(pool_size=(3, 3), strides=(2, 2)))

    # Flatten the output and add dense layers
    model.add(Flatten())
    model.add(Dense(4096, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(4096, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(num_classes, activation='softmax'))

    return model

# Mount Google Drive using the alternative method
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

# Set the path to your dataset in Google Drive
drive_path = '/content/drive/MyDrive/ColabNotebooks/Project/Dataset'
project_path = os.path.join(drive_path, '/content/drive/MyDrive/ColabNotebooks/Project/')
dataset_path = os.path.join(project_path, '/content/drive/MyDrive/ColabNotebooks/Project/Dataset')
train_data_dir = os.path.join(dataset_path, '/content/drive/MyDrive/ColabNotebooks/Project/Dataset/Train')
validation_data_dir = os.path.join(dataset_path, '/content/drive/MyDrive/ColabNotebooks/Project/Dataset/Validation')
test_data_dir = os.path.join(dataset_path, '/content/drive/MyDrive/ColabNotebooks/Project/Dataset/Test')

# Create necessary directories if they do not exist
for path in [project_path, dataset_path, train_data_dir, validation_data_dir, test_data_dir]:
    if not os.path.exists(path):
        os.makedirs(path)

# Image dimensions
img_width, img_height = 227, 227  # Adjusted for AlexNet

# Batch size
batch_size = 32

# Specify the number of classes
num_classes = 4  # Adjust based on your classification

# Create AlexNet model
alexnet_model = create_alexnet_model((img_width, img_height, 3), num_classes)

# Compile the model
alexnet_model.compile(optimizer=Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])

# Use ImageDataGenerator for data augmentation
train_datagen = ImageDataGenerator(rescale=1./255, shear_range=0.2, zoom_range=0.2, horizontal_flip=True)
test_datagen = ImageDataGenerator(rescale=1./255)

# Prepare the training dataset
train_generator = train_datagen.flow_from_directory(train_data_dir, target_size=(img_width, img_height),
                                                    batch_size=batch_size, class_mode='categorical', shuffle=True)

# Print information about the training dataset
print("Number of classes:", len(train_generator.class_indices))
print("Class indices:", train_generator.class_indices)
print("Number of training samples:", train_generator.samples)

# Prepare the validation dataset
validation_generator = test_datagen.flow_from_directory(validation_data_dir, target_size=(img_width, img_height),
                                                        batch_size=batch_size, class_mode='categorical')

# Print information about the validation dataset
print("Number of validation samples:", validation_generator.samples)

# Prepare the testing dataset
test_generator = test_datagen.flow_from_directory(test_data_dir, target_size=(img_width, img_height),
                                                  batch_size=batch_size, class_mode='categorical')

# Print information about the testing dataset
print("Number of test samples:", test_generator.samples)

# Train the model
history = alexnet_model.fit(train_generator, epochs=5, validation_data=validation_generator)

# Evaluate the model on the test set
test_loss, test_acc = alexnet_model.evaluate(test_generator)
print(f'Test accuracy: {test_acc}')

# Test the model on multiple sample images
test_folder_path = test_data_dir
test_images = [f for f in os.listdir(test_folder_path) if os.path.isfile(os.path.join(test_folder_path, f))]

for test_image_name in test_images:
    test_image_path = os.path.join(test_folder_path, test_image_name)
    test_image = tf.keras.preprocessing.image.load_img(test_image_path, target_size=(img_width, img_height))
    test_image = tf.keras.preprocessing.image.img_to_array(test_image)
    test_image = np.expand_dims(test_image, axis=0)
    test_image /= 255.0

    # Make a prediction
    prediction = alexnet_model.predict(test_image)

    # Map the prediction to the corresponding class
    classes = train_generator.class_indices
    predicted_class = list(classes.keys())[list(classes.values()).index(prediction.argmax())]

    # Display the result for each image
    print(f"Image: {test_image_name}, Predicted Class: {predicted_class}")
```

```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.applications import DenseNet201
from tensorflow.keras.preprocessing import image

# Mount Google Drive using the alternative method
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

# Set the path to your dataset in Google Drive
drive_path = '/content/drive/MyDrive/ColabNotebooks/Project/Dataset'
project_path = os.path.join(drive_path, '/content/drive/MyDrive/ColabNotebooks/Project/')
dataset_path = os.path.join(project_path, '/content/drive/MyDrive/ColabNotebooks/Project/Dataset')
train_data_dir = os.path.join(dataset_path, '/content/drive/MyDrive/ColabNotebooks/Project/Dataset/Train')
validation_data_dir = os.path.join(dataset_path, '/content/drive/MyDrive/ColabNotebooks/Project/Dataset/Validation')
test_data_dir = os.path.join(dataset_path, '/content/drive/MyDrive/ColabNotebooks/Project/Dataset/Test')

# Create necessary directories if they do not exist
for path in [project_path, dataset_path, train_data_dir, validation_data_dir, test_data_dir]:
    if not os.path.exists(path):
        os.makedirs(path)

# Image dimensions
img_width, img_height = 224, 224  # Adjusted for DenseNet201

# Batch size
batch_size = 32

# Specify the number of classes
num_classes = 4  # Adjust based on your classification

# Load pre-trained DenseNet201 model
base_model = DenseNet201(weights='imagenet', include_top=False, input_shape=(img_width, img_height, 3))

# Add custom layers on top of DenseNet201
model = Sequential()
model.add(base_model)
model.add(GlobalAveragePooling2D())
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(num_classes, activation='softmax'))

# Compile the model
model.compile(optimizer=Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])

# Use ImageDataGenerator for data augmentation
train_datagen = image.ImageDataGenerator(preprocessing_function=tf.keras.applications.densenet.preprocess_input, shear_range=0.2, zoom_range=0.2, horizontal_flip=True)
test_datagen = image.ImageDataGenerator(preprocessing_function=tf.keras.applications.densenet.preprocess_input)

# Prepare the training dataset
train_generator = train_datagen.flow_from_directory(train_data_dir,
                                                    target_size=(img_width, img_height),
                                                    batch_size=batch_size,
                                                    class_mode='categorical',
                                                    shuffle=True)

# Print information about the training dataset
print("Number of classes:", len(train_generator.class_indices))
print("Class indices:", train_generator.class_indices)
print("Number of training samples:", train_generator.samples)

# Prepare the validation dataset
validation_generator = test_datagen.flow_from_directory(validation_data_dir,
                                                        target_size=(img_width, img_height),
                                                        batch_size=batch_size,
                                                        class_mode='categorical')

# Print information about the validation dataset
print("Number of validation samples:", validation_generator.samples)

# Prepare the testing dataset
test_generator = test_datagen.flow_from_directory(test_data_dir,
                                                  target_size=(img_width, img_height),
                                                  batch_size=batch_size,
                                                  class_mode='categorical')

# Print information about the testing dataset
print("Number of test samples:", test_generator.samples)

# Train the model
history = model.fit(train_generator, epochs=5, validation_data=validation_generator)

# Evaluate the model on the test set
test_loss, test_acc = model.evaluate(test_generator)
print(f'Test accuracy: {test_acc}')

# Test the model on multiple sample images
test_folder_path = test_data_dir
test_images = [f for f in os.listdir(test_folder_path) if os.path.isfile(os.path.join(test_folder_path, f))]

for test_image_name in test_images:
    test_image_path = os.path.join(test_folder_path, test_image_name)
    test_image = image.load_img(test_image_path, target_size=(img_width, img_height))
    test_image = image.img_to_array(test_image)
    test_image = np.expand_dims(test_image, axis=0)
    test_image = tf.keras.applications.densenet.preprocess_input(test_image)

    # Make a prediction
    prediction = model.predict(test_image)

    # Map the prediction to the corresponding class
    classes = train_generator.class_indices
    predicted_class = list(classes.keys())[list(classes.values()).index(prediction.argmax())]

    # Display the result for each image
    print(f"Image: {test_image_name}, Predicted Class: {predicted_class}")
```
