AlexNet

```python
# ==========================================================
# ALEXNET MODEL DEFINITION
# Define the AlexNet architecture for image classification.
# ==========================================================

def create_alexnet_model(input_shape, num_classes):

    model = Sequential()

    # First convolution and pooling block
    model.add(
        Conv2D(
            96,
            (11, 11),
            strides=(4, 4),
            activation='relu',
            input_shape=input_shape
        )
    )
    model.add(
        MaxPooling2D(
            pool_size=(3, 3),
            strides=(2, 2)
        )
    )

    # Second convolution and pooling block
    model.add(Conv2D(256, (5, 5), activation='relu'))
    model.add(
        MaxPooling2D(
            pool_size=(3, 3),
            strides=(2, 2)
        )
    )

    # Third convolution layer
    model.add(Conv2D(384, (3, 3), activation='relu'))

    # Fourth convolution layer
    model.add(Conv2D(384, (3, 3), activation='relu'))

    # Fifth convolution and pooling block
    model.add(Conv2D(256, (3, 3), activation='relu'))
    model.add(
        MaxPooling2D(
            pool_size=(3, 3),
            strides=(2, 2)
        )
    )

    # Fully connected classification layers
    model.add(Flatten())
    model.add(Dense(4096, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(4096, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(num_classes, activation='softmax'))

    return model

# ==========================================================
# IMAGE PREPROCESSING PARAMETERS
# Configure image dimensions and training parameters.
# ==========================================================

# Image dimensions required by AlexNet
img_width, img_height = 227, 227

# Number of images processed per training iteration
batch_size = 32

# Number of dog weight classification categories
num_classes = 4


# ==========================================================
# MODEL ARCHITECTURE
# Create and compile the AlexNet classifier.
# ==========================================================

# Create AlexNet model
alexnet_model = create_alexnet_model(
    (img_width, img_height, 3),
    num_classes
)

# Compile model
alexnet_model.compile(
    optimizer=Adam(learning_rate=0.0001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)


# ==========================================================
# IMAGE DATA GENERATORS
# Apply preprocessing and data augmentation techniques.
# ==========================================================

# Data augmentation for training images
train_datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)

# Rescale validation and testing images
test_datagen = ImageDataGenerator(
    rescale=1./255
)


# ==========================================================
# DATASET LOADING
# Load training, validation, and testing datasets.
# ==========================================================

# Prepare training dataset
train_generator = train_datagen.flow_from_directory(
    train_data_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=True
)

# Display training dataset information
print("Number of classes:", len(train_generator.class_indices))
print("Class indices:", train_generator.class_indices)
print("Number of training samples:", train_generator.samples)

# Prepare validation dataset
validation_generator = test_datagen.flow_from_directory(
    validation_data_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='categorical'
)

print("Number of validation samples:",
      validation_generator.samples)

# Prepare testing dataset
test_generator = test_datagen.flow_from_directory(
    test_data_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='categorical'
)

print("Number of test samples:",
      test_generator.samples)


# ==========================================================
# MODEL TRAINING
# Train the AlexNet model using augmented image data.
# ==========================================================

history = alexnet_model.fit(
    train_generator,
    epochs=5,
    validation_data=validation_generator
)


# ==========================================================
# MODEL EVALUATION
# Evaluate model performance using the test dataset.
# ==========================================================

test_loss, test_acc = alexnet_model.evaluate(
    test_generator
)

print(f'Test accuracy: {test_acc}')


# ==========================================================
# SAMPLE IMAGE PREDICTIONS
# Run inference on test images and display predicted
# dog weight classifications.
# ==========================================================

test_folder_path = test_data_dir

test_images = [
    f for f in os.listdir(test_folder_path)
    if os.path.isfile(
        os.path.join(test_folder_path, f)
    )
]

for test_image_name in test_images:

    test_image_path = os.path.join(
        test_folder_path,
        test_image_name
    )

    test_image = tf.keras.preprocessing.image.load_img(
        test_image_path,
        target_size=(img_width, img_height)
    )

    test_image = tf.keras.preprocessing.image.img_to_array(
        test_image
    )

    test_image = np.expand_dims(
        test_image,
        axis=0
    )

    test_image /= 255.0

    # Generate prediction
    prediction = alexnet_model.predict(
        test_image
    )

    # Map prediction index to class label
    classes = train_generator.class_indices

    predicted_class = list(classes.keys())[
        list(classes.values()).index(
            prediction.argmax()
        )
    ]

    print(
        f"Image: {test_image_name}, "
        f"Predicted Class: {predicted_class}"
    )
```

```python
# ==========================================================
# DATASET CONFIGURATION
# Define dataset locations used for training, validation,
# and testing.
# ==========================================================

# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Set the path to your dataset on Google Drive
drive_path = '/content/drive/MyDrive/ColabNotebooks/Project/Dataset'
project_path = os.path.join(drive_path, '/content/drive/MyDrive/ColabNotebooks/Project/')
dataset_path = os.path.join(project_path, '/content/drive/MyDrive/ColabNotebooks/Project/Dataset')
train_data_dir = os.path.join(dataset_path, '/content/drive/MyDrive/ColabNotebooks/Project/Dataset/Train')
validation_data_dir = os.path.join(dataset_path, '/content/drive/MyDrive/ColabNotebooks/Project/Dataset/Validation')
test_data_dir = os.path.join(dataset_path, '/content/drive/MyDrive/ColabNotebooks/Project/Dataset/Test')

# Create necessary directories if they do not exist
for path in [project_path, dataset_path, train_data_dir,
             validation_data_dir, test_data_dir]:
    if not os.path.exists(path):
        os.makedirs(path)

# ==========================================================
# IMAGE PREPROCESSING PARAMETERS
# Configure image dimensions and training parameters.
# ==========================================================

# Image dimensions required by DenseNet201
img_width, img_height = 224, 224

# Number of images processed per training iteration
batch_size = 32

# Number of dog weight classification categories
num_classes = 4

# ==========================================================
# MODEL ARCHITECTURE
# Load a pretrained DenseNet201 backbone and build
# a custom transfer learning classifier.
# ==========================================================

# Load pretrained DenseNet201 feature extractor
base_model = DenseNet201(
    weights='imagenet',
    include_top=False,
    input_shape=(img_width, img_height, 3)
)

# Add custom classification layers
model = Sequential()
model.add(base_model)
model.add(GlobalAveragePooling2D())
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(num_classes, activation='softmax'))

# Compile the model
model.compile(
    optimizer=Adam(learning_rate=0.0001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# ==========================================================
# IMAGE DATA GENERATORS
# Apply DenseNet201 preprocessing and data augmentation.
# ==========================================================

# Data augmentation for training images
train_datagen = image.ImageDataGenerator(
    preprocessing_function=tf.keras.applications.densenet.preprocess_input,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)

# Preprocessing for validation and testing images
test_datagen = image.ImageDataGenerator(
    preprocessing_function=tf.keras.applications.densenet.preprocess_input
)

# ==========================================================
# DATASET LOADING
# Load training, validation, and testing datasets.
# ==========================================================

# Prepare training dataset
train_generator = train_datagen.flow_from_directory(
    train_data_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=True
)

# Display training dataset information
print("Number of classes:", len(train_generator.class_indices))
print("Class indices:", train_generator.class_indices)
print("Number of training samples:", train_generator.samples)

# Prepare validation dataset
validation_generator = test_datagen.flow_from_directory(
    validation_data_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='categorical'
)

# Display validation dataset information
print("Number of validation samples:",
      validation_generator.samples)

# Prepare testing dataset
test_generator = test_datagen.flow_from_directory(
    test_data_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='categorical'
)

# Display testing dataset information
print("Number of test samples:",
      test_generator.samples)

# ==========================================================
# MODEL TRAINING
# Train the DenseNet201 transfer learning model using
# augmented image data.
# ==========================================================

history = model.fit(
    train_generator,
    epochs=5,
    validation_data=validation_generator
)

# ==========================================================
# MODEL EVALUATION
# Evaluate model performance using the test dataset.
# ==========================================================

test_loss, test_acc = model.evaluate(
    test_generator
)

print(f'Test accuracy: {test_acc}')

# ==========================================================
# SAMPLE IMAGE PREDICTIONS
# Run inference on test images and display predicted
# dog weight classifications.
# ==========================================================

test_folder_path = test_data_dir

test_images = [
    f for f in os.listdir(test_folder_path)
    if os.path.isfile(
        os.path.join(test_folder_path, f)
    )
]

for test_image_name in test_images:

    test_image_path = os.path.join(
        test_folder_path,
        test_image_name
    )

    test_image = image.load_img(
        test_image_path,
        target_size=(img_width, img_height)
    )

    test_image = image.img_to_array(
        test_image
    )

    test_image = np.expand_dims(
        test_image,
        axis=0
    )

    test_image = tf.keras.applications.densenet.preprocess_input(
        test_image
    )

    # Generate prediction
    prediction = model.predict(
        test_image
    )

    # Map prediction index to class label
    classes = train_generator.class_indices

    predicted_class = list(classes.keys())[
        list(classes.values()).index(
            prediction.argmax()
        )
    ]

    # Display prediction results
    print(
        f"Image: {test_image_name}, "
        f"Predicted Class: {predicted_class}"
    )
```
